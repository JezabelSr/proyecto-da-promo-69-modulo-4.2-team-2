```python
df = pd.read_csv(os.path.join(RUTA, 'df_main.csv'))

df_generos = df[['tconst', 'genres']].copy()
df_generos['genres'] = df_generos['genres'].str.split(',')
df_generos = df_generos.explode('genres')
df_generos['genres'] = df_generos['genres'].str.strip()
df_generos = df_generos.dropna()

print(df_generos.shape)
print(df_generos['genres'].value_counts())
print(df_generos.head(10))
```

```python
df_generos.to_csv(os.path.join(RUTA, 'df_generos.csv'), index=False)
print("Guardado!")
```
